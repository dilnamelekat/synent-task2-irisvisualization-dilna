```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Iris.csv")
df.head()
df.info()
df.describe()
df['Species'].value_counts()
df.hist(figsize=(10,8))
plt.show()
sns.scatterplot(
    data=df,
    x='PetalLengthCm',
    y='PetalWidthCm',
    hue='Species'
)
plt.show()
avg_by_species = df.groupby('Species').mean()
avg_by_species.plot(kind='bar')
plt.show()
```
